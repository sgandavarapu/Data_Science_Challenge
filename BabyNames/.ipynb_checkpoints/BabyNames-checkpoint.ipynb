```python
import glob
import matplotlib.pyplot as plt
from pylab import rcParams
import collections
from collections import defaultdict
import operator

%matplotlib inline

all_states = glob.glob("C:\\Users\\SXG0748\\Documents\\Data_Science_Challenge\\BabyNames\\namesbystate\\*.txt")
Chosen_Year = '2013' #for question 3: What is the most gender ambiguous name in 2013? 1945? toggle it between 2013/1945.
Chosen_Year_threshold = 1980 #for question 4: Increase in popularity since 1980.

sumstate={}
sumyear={}
sumgender={}
sumname={}
sumnameyear={}
sumnameyear_yyyy={}

existing_keys = list()
sumgendername=defaultdict(list)
boynameratio=0.00
gender_ambig = list()
namecounts=[]

for infile in all_states:
    for line in open(infile):
        line = line.strip().split(',')
        state=line[0]
        #print state
        gender=line[1]
        year=line[2]
        name=line[3]
        count=int(line[4])
        namecounts.append(count)
        
    #basic summaries
        sumstate[state]=sumstate.get(state,0) + count
        sumyear[year]=sumyear.get(year,0) + count
        sumgender[gender] = sumgender.get(gender,0) + count
        nameyear=name+","+year
        
        sumnameyear[nameyear] = sumnameyear.get(nameyear,0)+count
    #Popularity Increase/Decrease    
        if int(year)>Chosen_Year_threshold:                   
            sumnameyear_yyyy[nameyear] = sumnameyear_yyyy.get(nameyear,0)+count
            
    #Most Popular Name
        sumname[name] = sumname.get(name,0) + count
        
    # Gender Ambigous Name
        if year == Chosen_Year:            
            if name not in existing_keys:
                sumgendername[name]=[0,0]
                existing_keys.append(name)
            if gender == 'M':
                sumgendername[name][0] = sumgendername.get(name,[])[0] + count #Boy name count
            else:
                sumgendername[name][1] = sumgendername.get(name,[])[1] + count # Girl name count


def popincdec(sumnameyear):
    #Additional processing for Popularity Inc/Dec
    #Function that defines increase and decrease in popularity for each name across years.
    df={}
    final_inc={}
    final_dec={}
    final_maxcount={}
    
    sorted_sumnameyear = sorted(sumnameyear.items(), key=operator.itemgetter(0), reverse=True) #sort by name
    oldname =  'null'
    max_count =0.001
    for key,values in sorted_sumnameyear:
        #print key
        #print values
        parts = key.strip().split(",") #name and year
        newname = parts[0] #person name
        df[parts[1]]=values #counts stored into year for each name
        if oldname<>newname:
            sorted_df = sorted(df.items(), key=operator.itemgetter(1),reverse=True) #sort by count value, 
            #print sorted_df
            for df_keys,df_values in sorted_df:
                max_year=int(df_keys) #find year in which max count turns up
                max_count=int(df_values)  # find max count
                break
            sorted_year_df = sorted(df.items(), key=operator.itemgetter(0)) #sort by year
            curr_min1=int(max_count) #initialize min before max
            curr_min2=int(max_count) #initialize min after max
            for df_keys,df_values in sorted_year_df: #implement logic for minimum bothe before and after maximum has been achieved
                if max_year>int(df_keys):
                    if curr_min1>df_values:
                        curr_min1=df_values
                if max_year<int(df_keys):
                    if curr_min2>df_values:
                        curr_min2=df_values
                perc_inc = round(100.0*(max_count-curr_min1)/float(curr_min1),3)
                perc_dec = round(100.0*(max_count-curr_min2)/float(max_count),3)
            #print oldname
            #print "max perc inc is " + str(perc_inc)
            #print "max perc dec is " + str(perc_dec)
            final_inc[oldname] = perc_inc
            final_dec[oldname] = perc_dec
            final_maxcount[oldname] = max_count
            
            df={} #initilize the dictionary with year as key and appearances as value for each name
            oldname = newname
        else:
            df[parts[1]]=values
            oldname = newname
    return final_inc,final_dec

final_inc,final_dec = popincdec(sumnameyear)
final_inc_yyyy,final_dec_yyyy = popincdec(sumnameyear_yyyy)

####### Print Results
rcParams['figure.figsize'] = 15, 6
#bins=[0, 5, 10, 20, 30, 40, 50, 100]
plt.hist(namecounts,color='green',alpha=0.8,bins=[0, 5, 10, 20, 30, 40, 50, 100])
#plt.xticks(range(len(bins)), bins)
plt.show()

for key in sorted(sumname, key=sumname.get, reverse=True):
    print "Most popular Baby Name is: " + key + " appearing " + str(sumname[key]) +" times \n"
    break

sorted_sumgendername = sorted(sumgendername.items(), key=operator.itemgetter(1),reverse=True) #sort by value, 
for keys,values in sorted_sumgendername:
    boynameratio=round(values[0] / float(values[0]+values[1]),3)
    if (boynameratio >0.45 and boynameratio <0.55):
        print "Top Gender ambigous name in " + Chosen_Year + " : " + keys + " used for boys:"+ str(values[0])+ " and for girls:" + str(values[1]) + " times \n"
        break

#Popularity after yyyy
sorted_final_inc_yyyy = sorted(final_inc_yyyy.items(), key=operator.itemgetter(1),reverse=True) #sort by value, 
for keys,values in sorted_final_inc_yyyy:
    inc_threshold= float(values)
    name_inc_yyyy=keys
    print "Name with maximum increase in popularity after " + str(Chosen_Year_threshold)  + " " + str(keys)
    print "max increase percentage after " + str(Chosen_Year_threshold)  + " : " +   str(values) +"\n"
    break                    
sorted_final_dec_yyyy = sorted(final_dec_yyyy.items(), key=operator.itemgetter(1),reverse=True) #sort by value, 
for keys,values in sorted_final_dec_yyyy:
    name_dec_yyyy=keys
    dec_threshold= float(values)
    print "Name with maximum decrease in popularity after " + str(Chosen_Year_threshold)  + " " +  str(keys)
    print "max decrease percentage after " + str(Chosen_Year_threshold)  + " : " +  str(values) +"\n"
    break 
    
    
#overall popularity
sorted_final_inc = sorted(final_inc.items(), key=operator.itemgetter(1),reverse=True) #sort by value, 
for keys,values in sorted_final_inc:
    if float(values)<inc_threshold:
        break 
    print "Overall, name with popularity increase more than " + name_inc_yyyy + " is " + str(keys)
    #print "max increase percentage :" + str(values)
                   
sorted_final_dec = sorted(final_dec.items(), key=operator.itemgetter(1),reverse=True) #sort by value, 
for keys,values in sorted_final_dec:
    if float(values)<dec_threshold:
        break
    print "Overall, names with popularity decrease more than " + name_dec_yyyy + " is " + str(keys)
    #print "max decrease percentage :" + str(values)
 
        
####### Print Charts
sumyear = collections.OrderedDict(sorted(sumyear.items()))
sumstate = collections.OrderedDict(sorted(sumstate.items()))


plt.bar(range(len(sumyear)), sumyear.values())
plt.xticks(range(len(sumyear)), sumyear.keys())
plt.xticks(rotation=90)
plt.title('Total Names Registered, By Year')
plt.subplots_adjust(left=0.1, right=1.1, top=0.9, bottom=0.1)
plt.show()

plt.bar(range(len(sumstate)), sumstate.values())
#plt.xticks(range(len(sumstate)), sumstate.keys())
plt.tick_params(axis='x', labelsize=8)
plt.title('Total Names Registered, By State')
plt.show()

plt.pie(sumgender.values(),labels=sumgender.keys())
plt.axis('equal')
plt.title('Total Names Registered, By Gender')
plt.show()
```
